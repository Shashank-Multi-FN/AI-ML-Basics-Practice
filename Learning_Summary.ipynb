Supervised Learning (Short Summary)

Supervised learning is when we train a model using labeled data.
That means the correct answers are already given.
The model learns by comparing its predictions with the correct output and improves.
Examples: spam detection, marks prediction, house price prediction.

Unsupervised Learning (Short Summary)

Unsupervised learning is when we train a model using unlabeled data.
Here, the model finds patterns or groups by itself.
Examples: customer grouping, clustering similar products, market segmentation.
