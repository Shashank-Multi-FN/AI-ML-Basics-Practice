AI/ML Important Terms (15)

1. Artificial Intelligence (AI) – Making machines perform tasks that normally need human intelligence.
2. Machine Learning (ML) – A part of AI where machines learn from data.
3. Dataset – A collection of data used to train or test a model.
4. Feature – An individual input variable in data.
5. Label – The output value that a model tries to predict.
6. Training Data – Data used to teach the model.
7. Testing Data – Data used to check model performance.
8. Overfitting – When a model learns training data too well and fails on new data.
9. Underfitting – When a model is too simple to learn data patterns.
10. Accuracy – Percentage of correct predictions.
11. Precision – How many predicted positives are actually correct.
12. Recall – How many actual positives are correctly identified.
13. Supervised Learning – Learning with labeled data.
14. Unsupervised Learning – Learning without labeled data.
15. Model – A mathematical representation learned from data.
