```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [78, 85, 67, 90, 74]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks Bar Chart")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(df["Student"], df["Marks"], marker="o")
plt.title("Student Marks Line Chart")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["Marks"], bins=5)
plt.title("Marks Distribution Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Student", y="Marks", data=df)
plt.title("Seaborn Barplot of Student Marks")
plt.show()
```
